```python
from sklearn.datasets import load_iris
```

```python
data_dict = load_iris()
```

```python
# df = pd.DataFrame(data_dict["data"],columns = data_dict["feature_names"])
df = pd.DataFrame(data_dict.data,columns = data_dict.feature_names)
```

```python
df
```

```python
# Normalization 
scaler = MinMaxScaler()
scaler.fit(df)
df = scaler.transform(df)
```

```python
df = pd.DataFrame(df)
```

```python
df.columns = data_dict.feature_names
df.columns
```

```python
df["species"] = data_dict.target
```

```python
df["species"] = df["species"].replace({0:"setosa",1:"vericolor",2:"virginica"})
```

```python
df
```

```python
x = df.drop("species",axis = 1)
```

```python
x
```

```python
y = data_dict.target
```

```python
sns.scatterplot(x = df["sepal length (cm)"],y = df["sepal width (cm)"], hue=df["species"])
plt.title("Distribution of species")
```

```python
sns.scatterplot(x = df["sepal length (cm)"],y = df["petal length (cm)"], hue = df["species"])
```

```python
kmeans_model = KMeans(n_clusters = 3,random_state = 11)
```

```python
kmeans_model.fit(x)
```

```python
y_pred = kmeans_model.predict(x)
```

```python
y_pred   ## 0: Setosa , 1: Versicolor, 2: Virginica
```

```python
y  ## 0: Setosa , 1: Versicolor, 2: Virginica
```

```python
df_new = x.copy()
```

```python
df_new
```

```python
df_new["Species_label"] = df["species"]
```

```python
df_new["Target"] = data_dict.target
```

```python
df_new
```

```python
df_new["K-Means_pred_label"] = y_pred
```

```python
df_new
```

```python
df_new["Kmeans_pred_new_label"] = df_new["K-Means_pred_label"].replace({0:"Setosa",1:"Versicolor",2:"Virginica"})
```

```python
df_new
```

```python
hue_order = ["Setosa","Versicolor","Virginica"]

plt.figure(figsize=(20,8))

# First subplot
plt.subplot(1, 2, 1)
sns.scatterplot(
    data=df_new,
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="Species_label",
    hue_order=hue_order
)
plt.title("Original Labels")

# Second subplot
plt.subplot(1, 2, 2)
sns.scatterplot(
    data=df_new,
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="Kmeans_pred_new_label",
    hue_order=hue_order
)
plt.title("KMeans Predicted Labels")

plt.show()
```

```python
# Elbow

inertia_list = []
k_val = range(1,7)
for i in k_val:
    model = KMeans(n_clusters=i)
    model.fit(x)
#     inertia = model.inertia_
    inertia_list.append(model.inertia_)
    
```

```python
inertia_list
```

```python
plt.plot(k_val,inertia_list)
plt.xlabel("Number of Cluster")
plt.ylabel("Inertia")
plt.title("Elbow Method")
```

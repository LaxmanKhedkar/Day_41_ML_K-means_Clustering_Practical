```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
data_dict = load_iris()
```

```python
df = pd.DataFrame(data_dict["data"],columns=data_dict["feature_names"])
```

```python
df
```

```python
scaler = MinMaxScaler()
scaler.fit(df)
df = scaler.transform(df)
```

```python
df
```

```python
df = pd.DataFrame(df)
```

```python
df
```

```python
df.columns = data_dict.feature_names
```

```python
df["species"] = data_dict["target"]
```

```python
df
```

```python
df["species"] = data_dict.target
```

```python
data_dict.target_names
```

```python
df["species"] = df["species"].replace({0:"Setosa",1:"Versicolor",2:"Virginica"})
```

```python
df
```

```python
x = df.drop("species", axis = 1)
```

```python
x
```

```python
y = data_dict.target
```

```python
sns.scatterplot(df["sepal length (cm)"],df["sepal width (cm)"], hue=df["species"])
```

```python
sns.scatterplot(df["sepal length (cm)"],df["petal length (cm)"], hue=df["species"])
```

```python
kmeans_model = KMeans(n_clusters=3, random_state=11)
```

```python
kmeans_model.fit(x)
```

```python
y_pred = kmeans_model.predict(x)
```

```python
y_pred ## 0: Setosa, 1: Versicolor, 2: Verginica ## The labels are assigened randomly
```

```python
y ## 0: Setosa , 1: Versicolor, 2: Virginica
```

```python
df_new = x.copy()
```

```python
df_new
```

```python
df_new["Species_label"] = df["species"]
```

```python
df_new["Target"] = data_dict.target
```

```python
df_new
```

```python
df_new["Kmeans_Pred_Label"] = y_pred
```

```python
df_new
```

```python
df_new["Kmeans_pred_new_label"] = df_new["Kmeans_Pred_Label"].replace({0:"Setosa",1:"Versicolor",2:"Virginica"})
```

```python
df_new  #({0:"Setosa",1:"Versicolor",2:"Virginica"})
```

```python
hue_order = ["Setosa","Versicolor","Virginica"]
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
sns.scatterplot(df_new["sepal length (cm)"],df_new["petal length (cm)"], hue=df_new["Species_label"],hue_order=hue_order)
plt.subplot(1,2,2)
sns.scatterplot(df_new["sepal length (cm)"],df_new["petal length (cm)"], hue=df_new["Kmeans_pred_new_label"],hue_order=hue_order)
```

```python
inertia_list = []
k_val = range(1,7)
for i in k_val:
    model = KMeans(n_clusters=i)
    model.fit(x)
#     inertia = model.inertia_
    inertia_list.append(model.inertia_)
    
```

```python
inertia_list
```

```python
plt.plot(k_val,inertia_list)
plt.xlabel("Number of Cluster")
plt.ylabel("Inertia")
plt.title("Elbow Method")
```
